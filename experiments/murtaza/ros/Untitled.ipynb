```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.load('training_buffer.npy')
```

```python
subtracted = [val['subtracted'] for val in x]
```

```python
gravity = [val['gravity'] for val in x]
applied = [val['applied'] for val in x]
observed = [val['observed'] for val in x]
```

```python
plt.plot(gravity)
```

```python
plt.plot(observed)
```

```python
plt.plot(applied)
```

```python
def get_ith_joint_info(i, data):
    return [obs[i] for obs in data]
```

```python
t = 1000
dt = 100
j = 1
for j in range(7):
    applied_j0 = get_ith_joint_info(j, applied)[t:t+dt]
    gravity_j0 = get_ith_joint_info(j, gravity)[t:t+dt]
    obs_j0 = get_ith_joint_info(j, observed)[t:t+dt]
    sub_j0 = get_ith_joint_info(j, subtracted)[t:t+dt]
    time = np.arange(len(applied_j0))
    plt.figure()
    plt.plot(time, applied_j0, time, obs_j0, time, gravity_j0, time, sub_j0)
    plt.legend(['applied', 'obs', 'gravity', 'sub'], loc='upper right')
```

```python
plt.plot(subtracted)
```

```python
def display_ith_joint_info(i, data):
    joint = [obs[i] for obs in data]
    plt.close()
    plt.plot(joint)
    plt.show()
```

```python
display_ith_joint_info(1, subtracted)
```
