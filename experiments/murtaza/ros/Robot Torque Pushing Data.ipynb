```python
def get_ith_joint_info(i, data):
    return [obs[i] for obs in data]
```

```python
data_push = np.load('training_buffer_with_pushes.npy')
```

```python
def get_all_qualities(data):
    gravity = [val['gravity'] for val in data]
    applied = [val['applied'] for val in data]
    observed = [val['observed'] for val in data]
    subtracted = [val['subtracted'] for val in data]
    return [gravity, applied, observed, subtracted]
```

```python
gravity, applied, observed, subtracted = get_all_qualities(data_push)
```

```python
display_ith_joint_info(0, subtracted)
```

```python
t = 3000
dt = 5000
j = 1
for j in range(7):
    applied_j0 = get_ith_joint_info(j, applied)[t:t+dt]
    gravity_j0 = get_ith_joint_info(j, gravity)[t:t+dt]
    obs_j0 = get_ith_joint_info(j, observed)[t:t+dt]
    sub_j0 = get_ith_joint_info(j, subtracted)[t:t+dt]
    time = np.arange(len(applied_j0))
    plt.figure()
    plt.plot(time, applied_j0, time, obs_j0, time, gravity_j0, time, sub_j0)
    plt.legend(['applied', 'obs', 'gravity', 'sub'], loc='upper right')
```

```python
def plot_all(t, dt, applied, gravity, observed, subtracted):
    t = t
    dt = dt
    for j in range(7):
        applied_j0 = get_ith_joint_info(j, applied)[t:t+dt]
        gravity_j0 = get_ith_joint_info(j, gravity)[t:t+dt]
        obs_j0 = get_ith_joint_info(j, observed)[t:t+dt]
        sub_j0 = get_ith_joint_info(j, subtracted)[t:t+dt]
        time = np.arange(len(applied_j0))
        plt.figure()
        plt.plot(time, applied_j0, time, obs_j0, time, gravity_j0, time, sub_j0)
        plt.legend(['applied', 'obs', 'gravity', 'sub'], loc='lower left')
```
